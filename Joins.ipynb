```python
A = pd.DataFrame({'key': [0,1,2,3], 'A': ['A0', 'A1', 'A2', 'A3']})
A
```

```python
B = pd.DataFrame({'key': [0,1,2,4],'B': ['B0', 'B1', 'B2','B3']})
B
```

```python
A.join(B, how="outer", on= "key", lsuffix='_A', rsuffix='_B')
```

```python
inner_merge = pd.merge(A, B, on='key', how="inner")
inner_merge
```

```python
outer_merge = pd.merge(A, B, on='key', how="outer")
outer_merge
```

```python
left_merge = pd.merge(A, B, on='key', how="left")
left_merge
```

```python
right_merge = pd.merge(A, B, on='key', how="right")
right_merge
```
